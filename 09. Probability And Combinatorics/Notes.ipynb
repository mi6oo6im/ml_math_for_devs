```python
#imports
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

Live Demo   

```python
results = np.random.randint(1,7, size = 10_000)
```

```python
plt.figure(figsize=(40 , 5))
plt.plot(results)
plt.xlabel("# trial")
plt.ylabel("Result")
plt.show()
```

```python
counts = Counter(results)
```

```python
counts
```

```python
plt.hist(results, bins= range(1, 8))
```

```python
plt.bar(counts.keys(), counts.values())
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```
